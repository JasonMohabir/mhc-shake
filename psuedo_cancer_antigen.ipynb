```python
#11.25.19 extract n-mer from mutated protein 

import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
DATA = "mini_challenge_data/"
mut_train = pd.read_csv(DATA + "mut_train.txt", sep="\t")
mut_test = pd.read_csv(DATA + "mut_test.txt", sep="\t")
```

```python
mut_test
```

```python
cols = ['patient_id','Amino_acids','Variant_Type','Variant_Classification','IMPACT','Hugo_Symbol','HGVSc','HGVSp','SWISSPROT']
mut_parsed = mut_test[cols]
mut_parsed = mut_parsed[~(mut_parsed.Variant_Classification.isin(
    ["RNA","Silent", "5'Flank", "Intron", "IGR", "3'UTR", "5'UTR"]))]
```

```python
mut_test = mut_parsed[mut_parsed.patient_id == 'TEST_0147']
mut_test = mut_test.fillna(0)
mut_test
```

```python
mut_filt = mut_test[~np.array(mut_test.SWISSPROT == 0)]
mut_filt = mut_filt[~np.array(mut_filt.HGVSp == 0)]
mut_filt
```

```python
def get_prot(swissprot):

    uni_prot = swissprot

    import httplib2 as http
    import json

    try:
        from urlparse import urlparse
    except ImportError:
        from urllib.parse import urlparse

    headers = {
            'Accept': 'application/rdf+xml',
    }

    uri = 'https://www.uniprot.org'
    path = '/uniprot/%s.fasta'%(uni_prot)
    target = urlparse(uri+path)
    method = 'GET'
    body = ''

    h = http.Http()

    response, content = h.request(
            target.geturl(),
            method,
            body,
            headers)

    output = open("%s.fasta"%(uni_prot), "w")
    output.write(content.decode("utf-8"))
    output.close()

    from Bio import SeqIO
    record = SeqIO.read("%s.fasta"%(uni_prot), "fasta")
    return(str(record.seq))
```

```python
ctr = 0
failed = 0 

for index,row in mut_filt.iterrows():
    mutation_index = int(''.join(list(filter(str.isdigit, row.HGVSp))))
    try:
        prot_seq = get_prot(row.SWISSPROT)
        mut_nucleotide = row.Amino_acids.split("/")[1]
        if mut_nucleotide == "*":
            new_seq = prot_seq[:mutation_index-1] + prot_seq[mutation_index:]
        else:
            new_seq = prot_seq[:mutation_index-1] + mut_nucleotide + prot_seq[mutation_index:]
        
        tumor_antigen = new_seq[mutation_index-1 -5 : mutation_index-1 + 5]
        
        if tumor_antigen == "":
            print(row.Hugo_Symbol,"Failed to get antigen")
            failed += 1
        else:
            print(row.Hugo_Symbol,tumor_antigen)
    except:
        print(row.Hugo_Symbol,"Failed to get antigen")
        failed += 1
    ctr += 1
    
```
